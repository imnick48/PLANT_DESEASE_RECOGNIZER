```python
import zipfile
import pickle
import tensorflow as tf
from tensorflow.keras import layers
```

```python
!kaggle datasets download vipoooool/new-plant-diseases-dataset -p dataset
```

```python
with zipfile.ZipFile('dataset/new-plant-diseases-dataset.zip', 'r') as zip_ref:
    zip_ref.extractall('dataset')
```

```python
train_dir='/content/dataset/New Plant Diseases Dataset(Augmented)/New Plant Diseases Dataset(Augmented)/train'
data=tf.keras.utils.image_dataset_from_directory(
    train_dir,
    image_size=(256,256),
    batch_size=32,
    shuffle=True,
    seed=42,
    validation_split=0.2,
    subset='training'
)
```

```python
model=tf.keras.Sequential([
    layers.Conv2D(32,kernel_size=(3,3),activation='relu',input_shape=(256,256,3)),
    layers.MaxPooling2D(pool_size=(2,2)),

    layers.Conv2D(64,kernel_size=(3,3),activation='relu'),
    layers.MaxPooling2D(pool_size=(2,2)),

    layers.Conv2D(128,kernel_size=(3,3),activation='relu'),
    layers.MaxPooling2D(pool_size=(2,2)),

    layers.Conv2D(256,kernel_size=(3,3),activation='relu'),
    layers.MaxPooling2D(pool_size=(2,2)),

    layers.Flatten(),
    layers.Dense(256,activation='relu'),
    layers.Dense(38,activation='softmax')
])
```

```python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
model.fit(data,epochs=10)
```

```python
test_dir='/content/dataset/New Plant Diseases Dataset(Augmented)/New Plant Diseases Dataset(Augmented)/valid'
data=tf.keras.utils.image_dataset_from_directory(
    test_dir,
    image_size=(256,256),
    batch_size=32,
    seed=42,
    subset='validation',
    validation_split=0.2
)
history=model.evaluate(data) #87% Accuracy
```

```python
tf.keras.models.save_model(model,'model.keras')
```
